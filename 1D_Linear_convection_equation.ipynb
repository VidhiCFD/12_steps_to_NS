```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
c = 1                            #convective speed in m/s
time = 3                         #total runtime simulation will run in sec
x = 10                           #total length of domain in m
nx = 100                         #number of points by which domain is discretized
nt = 50                          #number of timesteps i.e time discretization
dx = x/nx                        #size of 1 cell
dt = time/nt                     #size of one timestep
lam = (c*dt)/dx                  #for stability lam should be less than 1
```

```python
print(lam,dx,dt)
```

```python
u = np.ones((nt,nx))           #gives value at each point in domain at each timestep
```

```python
# defining initial condition
for i  in range(0,nx):
    x = i*dx
    if (x>= 0.5 and x < 1):
        u[0,i] = 2

    else:
        u[0,i] = 1
print(u[0,:])
    
```

```python
plt.figure(figsize = (10,5))
x = np.linspace(0,10,100)
plt.plot(x , u[0,:])
plt.show()
```

```python
''' now as initial condition is defined we now find solution for all times using 
    backward difference(BD) in space and FD in time'''

for t in range(0,nt-1):
    for i in range(1,nx):
        u[t+1,i] = u[t,i] - lam*(u[t,i] - u[t,i-1])
```

```python
plt.figure(figsize = (10,5))
x = np.linspace(0,10,100)
for t in range(0,nt,2):
    plt.plot(x , u[t,:])
plt.show()
```

```python
'''The above graph shows how the wave propogates through time but ideally it should
not distort as its a linear convection . profile should remain same as initial condition 
and it should just convect.
so the reason for distortion over here is discretization .its not proper.
we should refine mesh more to reduce the discretization error.  '''
```

```python
x=10
time=3
nx = 5000                       
nt = 3000                     
dx = x/nx                        
dt = time/nt                    
lam1 = (c*dt)/dx 
print(lam1,dx,dt)

u1 = np.ones((nt,nx))  

for i  in range(0,nx):
    x = i*dx
    if (x>= 0.5 and x < 1):
        u1[0,i] = 2
    else:
        u1[0,i] = 1

for t in range(0,nt-1):
    for i in range(1,nx):
        u1[t+1,i] = u1[t,i] - lam1*(u1[t,i] - u1[t,i-1])
```

```python
plt.figure(figsize = (10,5))
x = np.linspace(0,10,nx)
for t in range(0,nt,200):
    plt.plot(x , u1[t,:])
plt.show()
```
